
A sequence can be defined recursively by giving a formula that relates the U nth element to the its predecssor U n-1th element.

curr = 0.5[prev + a/(prev**(m-1))]

where a>0 and m>1

a is the base integer and m is the power that makes the number irrational.
so 
a**1/m

```python
seq=[1]
a=5
m=2
for i in range(15):
    seq.append(0.5*(seq[-1] + a/(seq[-1]**(m-1))))
```

```python
import matplotlib.pyplot as plt

plt.plot(seq)
```
